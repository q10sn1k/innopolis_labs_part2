```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('survey_results_public.csv', sep=',')
```

# Популярные Технологии

```python
languages = df['LanguageHaveWorkedWith'].str.split(';').explode().value_counts()
languages.plot(kind='bar', figsize=(15,7))
plt.title('Популярные языки программирования')
plt.xlabel('Язык')
plt.ylabel('Число респондентов')
plt.show()
```

# Распределение Ролей

```python
roles = df['DevType'].str.split(';').explode().value_counts()
roles.plot(kind='bar', figsize=(15,7))
plt.title('Распределение ролей разработчиков')
plt.xlabel('Роль')
plt.ylabel('Число респондентов')
plt.show()
```
